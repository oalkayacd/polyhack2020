```python
import csv, os, time, datetime, geojson
        
siteArr = []

with open('construction-site.xls.csv') as csvFile:
    csvReader = csv.reader(csvFile, delimiter=',')
    lineCount=0
    for row in csvReader:
        if lineCount!=0:
            if row[2]!= "" and row[3]!="":                                
                dateStart = row[8].split("/")
                dateEnd = row[9].split("/")
                try:
                    if int(dateStart[0]) < 30 and int(dateEnd[0]) < 30 and int(dateStart[1]) < 12 and int(dateEnd[1]) < 12:
                        timestamp1 = time.mktime(datetime.datetime.strptime(row[8],"%d/%m/%Y").timetuple())
                        timestamp2 = time.mktime(datetime.datetime.strptime(row[9],"%d/%m/%Y").timetuple())
                    
                        dict = {
                            "bpFrom" : row[2],
                            "bpTo" : row[3],
                            "distanceFrom" : row[4],
                            "distanceTo" : row[5],
                            "startDate" : timestamp1,
                            "endDate" : timestamp2
                        }
                        siteArr.append(dict)
                except:
                    continue
                    
                        
                                          
                

                

        lineCount += 1
        
bpArr = []        
        
with open('linie-mit-betriebspunkten.xls.csv') as csvFile:
    csvReader = csv.reader(csvFile, delimiter=',')
    lineCount=0
    for row in csvReader:
        if lineCount!=0:
            if row[2]!= "" and row[10]!="":
                latLng=row[10].split(',')
                dict = {
                    "abbr" : row[2],
                    "Lat" : latLng[0],
                    "Lng" : latLng[1]
                }
                
                bpArr.append(dict)
        lineCount += 1
        
newCsv = []
newCsv.append(["bpFrom", "bpTo", "bpFromLat", "geoshape"])
for elem in siteArr:
    bpFromCoord =[]
    bpToCoord = []
    for i in bpArr:
        if i.get("abbr") == elem.get("bpFrom"):
            bpFromCoord.append(i.get("Lat"))
            bpFromCoord.append(i.get("Lng"))
        elif i.get("abbr") == elem.get("bpTo"):
            bpToCoord.append(i.get("Lat"))
            bpToCoord.append(i.get("Lng"))
        elif i.get("abbr") == elem.get("bpFrom") and i.get("abbr") == elem.get("bpTo"):
            bpToCoord.append(i.get("Lat"))
            bpToCoord.append(i.get("Lng"))
    
    if bpFromCoord != [] and bpToCoord != []:
        #geoJsonObj = geojson.Feature(geometry=LineString([(bpFromCoord[0],bpToCoord[1], 0, elem.get("startDate")), (bpToCoord[0],  bpToCoord[1], 0, elem.get("endDate"))]), properties=)
        addedDict = {
                "type": "FeatureCollection",
                "features": [
                  {
                    "type": "Feature",
                    "properties": { "vendor":  "A",
                    "vol":20},
                    "geometry": {
                      "type": "LineString",
                      "coordinates": [
                        [bpFromCoord[0],bpToCoord[1], 0, elem.get("startDate")],
                        [ bpToCoord[0],  bpToCoord[1], 0, elem.get("endDate")]
                      ]
                    }
                  }
                ]
        }
        
        newCsv.append([elem.get("bpFrom"), elem.get("bpTo"),addedDict])

        
for elem in newCsv : print(elem)        
        
with open('newConstSite.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(newCsv)

    
        
        
        
        
        
        
```
