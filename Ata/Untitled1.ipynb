```python
import csv, os, time, datetime, geojson
        
siteArr = []

with open('construction-site.xls.csv') as csvFile:
    csvReader = csv.reader(csvFile, delimiter=',')
    lineCount=0
    for row in csvReader:
        if lineCount!=0:
            if row[2]!= "" and row[3]!="":                                
                dateStart = row[8].split("/")
                dateEnd = row[9].split("/")
                if row[8] != "" and row[9] != "":
                    timestamp1 = time.mktime(datetime.datetime.strptime(row[8],"%m/%d/%Y").timetuple())
                    timestamp2 = time.mktime(datetime.datetime.strptime(row[9],"%m/%d/%Y").timetuple())
                    
                    dict = {
                    "bpFrom" : row[2],
                    "bpTo" : row[3],
                    "distanceFrom" : row[4],
                    "distanceTo" : row[5],
                    "startDate" : timestamp1,
                    "endDate" : timestamp2
                    }
                    siteArr.append(dict)
                

        

        lineCount += 1
        
bpArr = []        
        
with open('linie-mit-betriebspunkten.xls.csv') as csvFile:
    csvReader = csv.reader(csvFile, delimiter=',')
    lineCount=0
    for row in csvReader:
        if lineCount!=0:
            if row[2]!= "" and row[10]!="":
                latLng=row[10].split(',')
                dict = {
                    "abbr" : row[2],
                    "Lat" : latLng[0],
                    "Lng" : latLng[1]
                }
                
                bpArr.append(dict)
        lineCount += 1        

newCsv = []
newCsv.append(["bpFrom", "bpTo","geoshape"])
for elem in siteArr:
    bpFromCoord =[]
    bpToCoord = []
    for i in bpArr:
        if i.get("abbr") == elem.get("bpFrom") and i.get("abbr") == elem.get("bpTo"):
            bpToCoord.append(i.get("Lat"))
            bpToCoord.append(i.get("Lng"))
            bpFromCoord.append(i.get("Lat"))
            bpFromCoord.append(i.get("Lng"))
        elif i.get("abbr") == elem.get("bpFrom"):
            bpFromCoord.append(i.get("Lat"))
            bpFromCoord.append(i.get("Lng"))
    for i in bpArr:
        if i.get("abbr") == elem.get("bpTo"):
            bpToCoord.append(i.get("Lat"))
            bpToCoord.append(i.get("Lng"))

    
    if bpFromCoord != [] and bpToCoord != []:
        geoJsonObj = geojson.Feature(geometry=geojson.LineString([(float(bpFromCoord[0]),float(bpToCoord[1]), 0, elem.get("startDate")), (float(bpToCoord[0]),  float(bpToCoord[1]), 0, elem.get("endDate"))]))
        featureCollection = geojson.FeatureCollection([geoJsonObj])
        print(featureCollection)

        """
        addedDict = {
                "type": "FeatureCollection",
                "features": [
                  {
                    "type": "Feature",
                    "properties": { "vendor":  "A",
                    "vol":20},
                    "geometry": {
                      "type": "LineString",
                      "coordinates": [
                        [bpFromCoord[0],bpToCoord[1], 0, elem.get("startDate")],
                        [ bpToCoord[0],  bpToCoord[1], 0, elem.get("endDate")]
                      ]
                    }
                  }
                ]
            
        }
        """
        
        newCsv.append([elem.get("bpFrom"), elem.get("bpTo"),featureCollection])
              
        
with open('newConstSite.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(newCsv)

    
        
        
        
        
        
        
```
