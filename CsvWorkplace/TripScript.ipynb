```python
import csv, os, time, datetime, geojson
        
siteArr = []

with open('construction-site.xls.csv') as csvFile:
    csvReader = csv.reader(csvFile, delimiter=',')
    lineCount=0
    for row in csvReader:
        if lineCount!=0:
            if row[2]!= "" and row[3]!="":                                
                dateStart = row[8].split("/")
                dateEnd = row[9].split("/")
                if row[8] != "" and row[9] != "":
                    timestamp1 = time.mktime(datetime.datetime.strptime(row[8],"%m/%d/%Y").timetuple())
                    timestamp2 = time.mktime(datetime.datetime.strptime(row[9],"%m/%d/%Y").timetuple())
                    
                    dict = {
                    "bpFrom" : row[2],
                    "bpTo" : row[3],
                    "distanceFrom" : row[4],
                    "distanceTo" : row[5],
                    "startDate" : timestamp1,
                    "endDate" : timestamp2
                    }
                    siteArr.append(dict)
                

        

        lineCount += 1
        
bpArr = []        
        
with open('linie-mit-betriebspunkten.xls.csv') as csvFile:
    csvReader = csv.reader(csvFile, delimiter=',')
    lineCount=0
    for row in csvReader:
        if lineCount!=0:
            if row[2]!= "" and row[10]!="":
                latLng=row[10].split(',')
                dict = {
                    "abbr" : row[2],
                    "Lat" : latLng[1],
                    "Lng" : latLng[0]
                }
                
                bpArr.append(dict)
        lineCount += 1        

newCsv = []
newCsv.append(["bpFrom", "bpTo","geoshape"])
for elem in siteArr:
    bpFromCoord =[]
    bpToCoord = []
    for i in bpArr:
        if i.get("abbr") == elem.get("bpFrom") and i.get("abbr") == elem.get("bpTo"):
            bpToCoord.append(i.get("Lat"))
            bpToCoord.append(i.get("Lng"))
            bpFromCoord.append(i.get("Lat"))
            bpFromCoord.append(i.get("Lng"))
        elif i.get("abbr") == elem.get("bpFrom"):
            bpFromCoord.append(i.get("Lat"))
            bpFromCoord.append(i.get("Lng"))
    for i in bpArr:
        if i.get("abbr") == elem.get("bpTo"):
            bpToCoord.append(i.get("Lat"))
            bpToCoord.append(i.get("Lng"))

    
    if bpFromCoord != [] and bpToCoord != []:
        lineStrings = []
        lineStrings.append([(float(bpFromCoord[0]), float(bpFromCoord[1]), 0, elem.get("startDate")), ((float(bpFromCoord[0])+float(bpToCoord[0]))/2, (float(bpFromCoord[1])+float(bpToCoord[1]))/2, 0, (elem.get("startDate")+elem.get("endDate"))/2), (float(bpToCoord[0]), float(bpToCoord[1]), 0, elem.get("endDate"))])                  
        geoJsonObj = geojson.Feature(geometry=geojson.LineString(lineStrings))
        featureCollection = geojson.FeatureCollection([geoJsonObj])
        
        properties_dict = {
            "bpFrom" : elem.get("bpFrom"),
            "bpTo" : elem.get("bpTo"),
        }
        #Convert the dictionary to a list
        properties_list = list(zip(properties_dict.keys(),properties_dict.values()))
        #Loop over GeoJSON features and add the new properties
        for feat in featureCollection['features']:
            for i in range(len(properties_list)):
                feat ['properties'][properties_list[i][0]]=properties_list[i][1]
        
        print(featureCollection['features'][0])
        newCsv.append([elem.get("bpFrom"), elem.get("bpTo"),featureCollection['features'][0]])
              
        
with open('newConstSite.csv','w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(newCsv)

    
        
        
        
        
        
        
```
